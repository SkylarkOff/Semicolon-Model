```python
import pandas as pd
import numpy as np

# Konversi string ke datetime dengan zona waktu UTC
label['start_ts'] = pd.to_datetime(label['start_ts'].str.split().str[0], format='%m/%d/%y').dt.tz_localize('UTC')
label['end_ts'] = pd.to_datetime(label['end_ts'].str.split().str[0], format='%m/%d/%y').dt.tz_localize('UTC')


# Buat IntervalIndex untuk pencarian cepat
interval_index = pd.IntervalIndex.from_arrays(label['start_ts'], label['end_ts'], closed='both')

# Buat dictionary untuk mapping `pid`
pid_intervals = dict(zip(label['pid'], interval_index))

# Fungsi untuk menandai episode depresi
def check_depression(row):
    pid = row['pid']
    dt = row['datetime']
    if pid in pid_intervals and dt in pid_intervals[pid]:
        return 1
    return 0

# Terapkan ke dataset observasi
df['depression_episode'] = df.apply(check_depression, axis=1)

# Gabungkan data tanpa menambahkan kolom duplikat
merged_df = df.merge(label, on="pid", how="left", suffixes=("", "_drop"))

# Hapus kolom yang tidak diperlukan
merged_df = merged_df.drop(columns=['start_ts', 'end_ts', 'start_phq9', 'end_phq9', 'depression_episode_drop'], errors='ignore')

# Tampilkan hasil
merged_df.head()
```

```python
import pandas as pd

# Salin dataset untuk menjaga data asli tetap aman
df_fe = merged_df.copy()

# Contoh perhitungan 'contours_mean' jika menggunakan beberapa kolom kontur
contour_cols = ["contours_x", "contours_y"]  # Sesuaikan dengan nama kolom sebenarnya
df_fe["contours_mean"] = df_fe[contour_cols].mean(axis=1)

landmarks_cols = ["landmarks_x", "landmarks_y"]  # Sesuaikan dengan nama kolom sebenarnya
df_fe["landmarks_mean"] = df_fe[landmarks_cols].mean(axis=1)


# 1️⃣ Ekspresi Wajah Gabungan
df_fe["AU_Smile"] = (df_fe["AU06"] + df_fe["AU12"] + df_fe["smilingProbability"]) / 3
df_fe["AU_Sad"] = (df_fe["AU04"] + df_fe["AU15"] + df_fe["AU17"]) / 3

# 2️⃣ Rata-rata Keterbukaan Mata
df_fe["Eye_Open_Avg"] = (df_fe["leftEyeOpenProbability"] + df_fe["rightEyeOpenProbability"]) / 2

# 3️⃣ Gerakan Kepala (Head Movement)
df_fe["Head_Movement"] = abs(df_fe["pitch"]) + abs(df_fe["yaw"]) + abs(df_fe["roll"])

# 4️⃣ Fitur Geometris Wajah
df_fe["Facial_Structure"] = (df_fe["contours_mean"] + df_fe["landmarks_mean"]) / 2

# 5️⃣ Hapus fitur asli yang sudah digunakan dalam pembuatan fitur baru
drop_cols = ["AU06", "AU12", "AU04", "AU15", "AU17",
             "leftEyeOpenProbability", "rightEyeOpenProbability",
             "smilingProbability", "pitch", "yaw", "roll",
             "contours_mean", "landmarks_mean"]
df_fe = df_fe.drop(columns=drop_cols)

# Tampilkan 5 baris pertama hasil transformasi
print(df_fe.head())
```

```python
import pandas as pd

# Buang kolom 'pid' karena bukan fitur numerik
core = df_fe.drop(columns=['pid'])

# Hitung matriks korelasi
correlation_matrix = core.corr()

# Urutkan korelasi fitur terhadap target
target_corr = correlation_matrix["depression_episode"].sort_values(ascending=False)

# Tampilkan hasil korelasi dalam bentuk angka
print(target_corr)
```

```python
import pandas as pd
from scipy import stats  # Tambahkan ini untuk ANOVA

group_0 = df_fe[df_fe["depression_episode"] == 0]
group_1 = df_fe[df_fe["depression_episode"] == 1]

# Pilih hanya kolom numerik (exclude objek/string)
numeric_cols = df_fe.select_dtypes(include=['number']).columns.tolist()

# Pastikan 'depression_episode' tetap ada dalam daftar
numeric_cols.remove("depression_episode")

# Jalankan ANOVA hanya untuk kolom numerik
anova_results = {}
for col in numeric_cols:
    stat, p_value = stats.f_oneway(group_0[col], group_1[col])
    anova_results[col] = p_value

# Konversi ke DataFrame
anova_df = pd.DataFrame(anova_results.items(), columns=["Feature", "p-value"])
anova_df = anova_df.sort_values(by="p-value")

# Tampilkan fitur dengan p-value < 0.05 (signifikan)
significant_features_anova = anova_df[anova_df["p-value"] < 0.05]
print(significant_features_anova)
```
